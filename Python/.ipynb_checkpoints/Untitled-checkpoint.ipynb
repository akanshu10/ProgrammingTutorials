```python
import cv2

video = cv2.VideoCapture(0)
a = 1
while True:
    a = a+1
    check ,frame = video.read()
    print(check)
    print(frame)
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow("capture",gray)
    key = cv2.waitKey(1)
    
    if (key == ord('q')):
        break;
cv2.realse()
cv2.destroyAllWindows()
```
